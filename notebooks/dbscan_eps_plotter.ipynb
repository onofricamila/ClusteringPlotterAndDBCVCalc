```python
from sklearn.datasets import make_blobs
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#centers = [[2.75, -2], [2.5, 2], [1, -1]]
#stds = [0.2, 0.4, 0.45]

centers = [[1, 0.5], [2, 2], [1, -1]]
stds = [0.1, 0.4, 0.3]

X, labels_true = make_blobs(n_samples=1100, centers=centers, cluster_std=stds, random_state=0)
plt.scatter(X[:,0], X[:,1], c=[x for x in labels_true])

# X = pd.DataFrame(X)

# X.head()

# X.plot.scatter(x=0, y=1, c=[x for x in labels_true])
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
fig = plt.figure(figsize=(20, 20))
fig.subplots_adjust(hspace=.5, wspace=.2)
i = 1
for x in range(9, 0, -1):
    eps = 1/(10-x)
    db = DBSCAN(eps=eps, min_samples=10).fit(X)
    core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
    core_samples_mask[db.core_sample_indices_] = True
    labels = db.labels_
    
    print(eps)
    ax = fig.add_subplot(3, 3, i)
    ax.text(1, 4, "eps = {}".format(round(eps, 2)), fontsize=25, ha="center")
    # plt.scatter(X[:,0], X[:,1], c=[x for x in labels])
    sns.set(font_scale=3)

    sns.set(rc={'axes.facecolor':'lightgrey', 'figure.facecolor':'white'})
    s = sns.scatterplot(x=X[:,0], y=X[:,1], 
                    hue=labels,palette='rocket')
    sns.move_legend(s, "lower right", title='Grupo')
    # for legend text
    plt.setp(s.get_legend().get_texts(), fontsize='17') 

    # for legend title
    plt.setp(s.get_legend().get_title(), fontsize='20') 
    i += 1
```
